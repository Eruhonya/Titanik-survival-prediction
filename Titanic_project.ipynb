#### Ticket:
Ticket Code Breakdown for Titanic Dataset
1. PC (Pont et Chaussées)
What it means:
This code corresponds to first-class tickets purchased through the French agency Pont et Chaussées.

Who traveled:
Typically, passengers in first class, and also very wealthy individuals.

Example:
The ticket PC 17599 was held by Joseph Bruce Ismay, the director of White Star Line.

2. A/5 (or A/5 21171)
What it means:
This code refers to third-class tickets, often signifying more affordable tickets.

Who traveled:
Mainly immigrants, people traveling to America seeking a better life.

Example:
The ticket A/5 21171 was used by passengers traveling in the third-class section.

3. C.A. (Contract Agreement)
What it means:
This code denotes group or family tickets, often purchased for several people traveling together.

Who traveled:
Families or groups, typically immigrants heading to America.

Example:
For example, ticket C.A. 2677 was used by members of a family traveling together.

4. SOTON/OQ (Southampton to Queenstown)
What it means:
These tickets were for passengers boarding Titanic in Southampton, with a destination in Queenstown (modern Cork, Ireland).

Who traveled:
Mostly third-class passengers, primarily immigrants.

Example:
SOTON/OQ 392076 is one of such tickets used by passengers traveling to America.

5. W.E.P. (White Star Line)
What it means:
W.E.P. is a code for first-class tickets purchased through the official White Star Line service.

Who traveled:
Passengers traveling in first class, often from the wealthier classes of society.

Example:
The ticket W.E.P. 32110 was held by a first-class passenger traveling with White Star Line.

6. Other Ticket Codes
S. (Saloon) — This typically represents first-class passengers traveling in the luxurious saloon section of the ship.

L1, L2, etc. — These codes may represent different classes or cabin numbers, though they are not as common in the accessible data, but could be useful for further clarification.

```python
# Functions

# Encoder One-hot encoding(fit, transform)
def get_OHE(df, drop_param=True):
    one_enc = OneHotEncoder(
        top_categories=None,
        variables=['sex','embarked'],
        drop_last=drop_param)

    one_enc.fit(X_train)
    X_df_OHE = one_enc.transform(df)
    return X_df_OHE
    
# Encoder ordinal encoding(fit, transform)
def get_ordinal(df): 
    ord_enc= OrdinalEncoder(
        encoding_method='arbitrary')

    ord_enc.fit(X_train)
    X_df_Ord = ord_enc.transform(df)
    return X_df_Ord

# Count encoding(fit, transform)
def get_count(df): 
    count_enc = CountFrequencyEncoder(
        encoding_method='count')

    count_enc.fit(X_train)
    X_df_Count = count_enc.transform(df)
    return X_df_Count

# fit Logistic regresssion model
def run_logistic(X_train,X_test, y_train, y_test):
    scaler = StandardScaler()
    scaler.fit(X_train)
    X_train_scaled = scaler.transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    logit = LogisticRegression(random_state=0, C=0.01, max_iter=100)

    logit.fit(X_train_scaled, y_train)

    print("Train set")
    pred = logit.predict_proba(X_train_scaled)
    print("Logistic regression roc_auc: {}".format(roc_auc_score(y_train, pred[:, 1])))

    print()
    print("Test set")
    pred = logit.predict_proba(X_test_scaled)
    print("Logistic regression roc_auc: {}".format(roc_auc_score(y_test, pred[:, 1])))

# fit RandomForests model
def run_randomForests(X_train, X_test, y_train,y_test):
    rf = RandomForestClassifier(n_estimators=50,random_state=39, max_depth=3)
    rf.fit(X_train, y_train)

    print("Train set")
    pred = rf.predict_proba(X_train)
    print("Random forests roc_auc: {}".format(roc_auc_score(y_train, pred[:, 1])))

    print()
    print("Test set")
    pred = rf.predict_proba(X_test)
    print("Random forests roc_auc: {}".format(roc_auc_score(y_test, pred[:, 1])))

# Distribution plot
def diagnostic_plots(df, variable):
    plt.figure(figsize=(15,6))
    plt.subplot(1,2,1)
    df[variable].hist(bins=30)
    plt.title(variable)

    plt.subplot(1,2,2)
    stats.probplot(df[variable], dist='norm', plot=plt)
    plt.title(variable)
    plt.show()
# Confusion matrix
def plot_custom_confusion_matrix(y_true, y_pred, labels=None, normalize=False, title='Confusion Matrix'):
    cm = confusion_matrix(y_true, y_pred, normalize='true' if normalize else None)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt=".2f" if normalize else "d",
                cmap="Blues", xticklabels=labels, yticklabels=labels)
    plt.xlabel('False')
    plt.ylabel('True')
    plt.title(title)
    plt.tight_layout()
    plt.show()

# Calculation of metrics and appending to the list
def metrics_gridsearch(model_name, y_test, y_train, y_pred_test,y_proba_test,y_pred_train, y_proba_train, results_df):

    f1_test = round(f1_score(y_test, y_pred_test), 4)
    acc_test = round(accuracy_score(y_test, y_pred_test), 4)
    roc_test = round(roc_auc_score(y_test, y_proba_test), 4)

    f1_train = round(f1_score(y_train, y_pred_train), 4)
    acc_train = round(accuracy_score(y_train, y_pred_train), 4)
    roc_train = round(roc_auc_score(y_train, y_proba_train), 4)
    results_df[model_name + '_test'] = [f1_test, acc_test, roc_test]
    results_df[model_name + '_train'] = [f1_train, acc_train, roc_train]
    return results_df

# Final Summary
def plot_overfit_with_exact_labels(results_df, metric_name):
    df = results_df.T

    df_metric = df[[metric_name]].copy()
    df_metric['model'] = df_metric.index
    df_metric['type'] = df_metric['model'].apply(lambda x: 'train' if 'train' in x else 'test')
    df_metric['model_clean'] = df_metric['model'].apply(lambda x: x.replace('_train', '').replace('_test', ''))

    pivot = df_metric.pivot(index='model_clean', columns='type', values=metric_name).fillna(0)
    pivot['overfit_%'] = ((pivot['train'] - pivot['test']) / pivot['train']) * 100

    fig, ax = plt.subplots(figsize=(10, 6))
    train_bars = ax.barh(pivot.index, pivot['train'], color='orange', label='Train')
    test_bars = ax.barh(pivot.index, pivot['test'], color='steelblue', label='Test')
    for i, model in enumerate(pivot.index):
        train_val = pivot.loc[model, 'train']
        test_val = pivot.loc[model, 'test']
        diff = pivot.loc[model, 'overfit_%']
        ax.text(train_val - 0.01, i - 0.15, f'{train_val*100:.1f}%', color='purple', ha='right', fontsize=12)
        ax.text(test_val - 0.01, i + 0.05, f'{test_val*100:.1f}%', color='blue', ha='right', fontsize=12)
        ax.text(max(train_val, test_val) + 0.02, i, f'{diff:.1f}%', color='darkred', ha='left', va='bottom', fontsize=12)

    ax.set_xlabel(metric_name)
    ax.set_title(f'{metric_name}: Train vs Test with % difference')
    ax.set_xlim(0, 1.1)
    ax.legend()
    ax.grid(axis='x', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```

##### Thinking
<b>name</b> - Names do not have particular importance for survival. This variable may indicate family relationships between people, as reflected in the 'sibsp' and 'parch' variables, but since these variables do not contain missing values, there is essentially no need for it. A similar situation applies to titles. A title can indicate a person's status (and therefore their class and ticket price) as well as their gender, and to a lesser extent, their age.

<b>ticket</b> - The ticket number does not carry significant importance, but prefixes may indicate the passenger’s class and status, or a group (family) purchase. However, since these variables do not have missing values, there is no actual need to rely on the ticket number.

<b>body</b> - This variable directly indicates survival; if the body was found, it means the person perished. However, with 90% of the values missing, it may be difficult to recover this information. A missing value could also indicate either survival or death.

<b>boat</b> - The variable indicates which lifeboat the passenger was loaded onto if they were rescued. This variable is directly related to survival, but It has a large percentage of missing values, and not all survivors ended up in a lifeboat.

"cabin" - This variable shows the passenger's location by deck and cabin number. There is an important correlation between the location of the cabins and the decks, from A to G (from first class to third class). Consequently, the higher the deck, the closer it is to the lifeboats, which is associated with survival. Cabin numbers describe the location relative to the bow of the ship (by the tilt, it is possible to determine which part of the cabins sank first).This variable is also correlated with the ticket price and class (but the variables do not have missing values). It can also be assumed that not all passengers were in their cabins; they may have been in others' cabins or in different parts of the ship. However, this variable has a large number of missing values, which will significantly impact the quality of the model.

"home.dest" - This variable shows the place of departure or destination. Based on this variable, it is possible to roughly determine if the person was an immigrant (which could suggest the approximate class of the cabin). However, the class variable does not contain missing values, while this variable has a high cardinality and over 40% missing values, which could significantly impact the model.

```python
X_train, X_test, y_train, y_test = train_test_split(
    engindata.drop(['survived'], axis=1),
    engindata['survived'],
    test_size=0.3,
    random_state=0
)

X_train.shape,X_test.shape
```

```python
Tree_Pipe = Pipeline(steps = [
    ('imputer_age', SklearnTransformerWrapper(transformer=IterativeImputer(estimator=ExtraTreesRegressor(n_estimators=100, random_state=0),
                                                                          initial_strategy='median',
                                                                          max_iter=10,
                                                                          imputation_order='ascending',
                                                                          n_nearest_features=None,
                                                                          skip_complete=True,
                                                                          random_state=0))),
    ('transformer', vt.YeoJohnsonTransformer(variables=['age','fare'])),
    ('cat_encoder', OrdinalEncoder()),
    ('Tree', GradientBoostingClassifier( n_estimators=30, min_samples_split=3, min_samples_leaf=5, max_depth=2, random_state=0))
])
```

```python
Model = Tree_Pipe.fit(X_train,y_train)
```

```python
X_train_proba = Model.predict_proba(X_train)
X_test_proba = Model.predict_proba(X_test)
```
